```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Exercise 1
X = np.random.normal(0,4,size=(10000))
Norm_p = pd.DataFrame(X)
```

```python
#Exercise 2
plt.hist(X,histtype='step')  # arguments are passed to np.histogram
plt.title("Histogram")
plt.show()
```

```python
#Exercise 3
Cum_Mean = Norm_p.expanding().mean()
```

```python
#Exercise 4
Cum_std_err = Norm_p.rolling(len(Norm_p),min_periods=2).std()
```

```python
# Exercise 5 Bare mean and std deviation?
Norm_P_mean = np.mean(Norm_p)
Norm_P_median = np.median(Norm_p)

# However, our distribution mean = 0, 
```

```python
#Exercise 6
x_vals = list(range(1,len(Cum_Mean)+1))

plt.plot(x_vals,Cum_Mean,'-')
```

```python
fig = plt.figure()
x = np.array(x_vals)[1:]
ye = np.array(Cum_Mean)[1:] # First value is excluded as standard error of a single point is Nan
y = [xe[0] for xe in ye]
yerr=np.array(Cum_std_err[1:len(x)+1][0])

plt.errorbar(x, y, yerr=yerr, label='both limits (default)',linewidth = 0.003)
plt.plot(x_vals,Cum_Mean,'-')

plt.legend(loc='lower right')
plt.plot([1,2,3,10000],[Norm_P_mean,Norm_P_mean,Norm_P_mean,Norm_P_mean])
plt.show()
```
